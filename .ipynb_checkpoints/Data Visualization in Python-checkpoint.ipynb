## Univariate plots

### Pandas and matplotlib

```python
medalist_weights.head(3)
```

```python
medalist_weights['Height'].plot(kind='hist')
plt.show()
```

```python
medalist_weights['Weight'].plot(kind='hist')
plt.show()
```

```python
# Boxplot
medalist_weights['Weight'].plot(kind='box')
plt.show()
```

```python
# Bar plot
cts = medalist_weights['Medal'].value_counts()
cts.plot(kind='bar')
plt.show()
```

### Seaborn

```python
# Bar plot
sns.countplot(x='Sex', data=medalist_weights)
plt.show()
```

```python
# Histogram using Seaborn
sns.displot(medalist_weights['Height'])
plt.show()
```

## Bivariate plots

```python
# Scatter plot
medalist_weights.plot(x='Weight', y='Height', kind='scatter')
plt.show()
```

```python
sns.regplot(x='Weight', y='Height', data=medalist_weights)
plt.show()
```

You can use a boxplot to compare one continuous and one categorical variable. However, you will be using the .boxplot() method instead of the .plot() method

```python
# Boxplot of the weight column by sex
medalist_weights.boxplot(column='Weight', by='Sex')
plt.show()
```

```python
# Boxplot of the height column by sex
medalist_weights.boxplot(column='Height', by='Sex')
plt.show()
```

```python
# Boxplot
sns.boxplot(x="Sex", y="Height", data=medalist_weights)
plt.show()
```
